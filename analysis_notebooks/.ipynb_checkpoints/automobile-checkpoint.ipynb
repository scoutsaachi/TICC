# Automobile Analysis
This is a notebook to generate the figures from the automobile data

```python
%matplotlib inline
```

```python
import pandas as pd
import datetime
import pickle
import matplotlib.pyplot as plt
```

Load the data

```python
IDENTIFIER = 205
path = "../car_data/%s" % IDENTIFIER # car data path here
df = pd.read_csv('%s/turnSensors.csv' % path, sep=",", header=None)
df.columns = ["Heading", "Velocity"]
dfLatLong = pd.read_csv('%s/latlong.csv' % path, sep=",", header=None)
dfLatLong.columns = ["Lat", "Long"]
dfLatLong[(dfLatLong["Lat"]<0)] = float('nan')
dfLatLong.fillna(inplace=True, method="ffill")
resultDf = pd.read_csv('%s/0.6/assign.out' % path, sep=",", header=None)
resultDf.columns=["Assignment"]
oldDf = pd.read_csv('%s/old/assign.out' % path, sep=",", header=None)
oldDf.columns=["Assignment"]
f = open('%s/0.6/motifs.pkl' % path, "rb")
motifs = pickle.load(f)
# print(motifs)
m = (6, 7, 0)
motifIndices = motifs[m]
print(sorted(list(motifIndices)))
```

```python
def collapse(df, gap=None):
    if gap is not None:
        indivFrame = df[gap[0]: gap[1]+1]
    else:
        indivFrame = df
    values = indivFrame.values.T.flatten()
    prev = values[0]
    result = [str(prev)]
    for v in values:
        if v != prev:
            result.append(str(v))
            prev = v
    return result
```

Check instances of the motif in TICC's assignment

```python
totalLen = oldDf.count().values[0]
collapseWindow = collapse(oldDf)
concatenated = ''.join([str(i) for i in collapseWindow])
concatMotif = ''.join([str(i) for i in m])

print("TICC all occurrences", concatenated.count(concatMotif))
correct = 0
motifFound = []
for start, end in motifIndices:
    result = collapse(oldDf, (start, end))
    resultStr = ''.join(result)
    if concatMotif in resultStr:
        correct += 1
        motifFound.append((start, end))
print("TICC found %s out of %s" % (correct, len(motifIndices)))
```

Motifs plotted on entire map

```python
colors = ["r", "b", "g", "orange", "c", "y", "magenta", "brown", "gray"]

useOffsets = False

plt.figure(1, figsize=[7,9])
for i in range(8):
    indices = resultDf["Assignment"] == i
    latLong = dfLatLong[indices]
    plt.plot(latLong["Long"].values, latLong["Lat"].values, "o", markersize=2, color="lightgray")
    
for i, gap in enumerate(motifIndices):
    latLong = dfLatLong[gap[0]:gap[1]+1]
    label = "CASC ONLY" if i == 0 else None
    plt.plot(latLong["Long"].values, latLong["Lat"].values, "*", markersize=4, color="blue", label=label)

for i, gap in enumerate(motifFound):
    latLong = dfLatLong[gap[0]:gap[1]+1]
    label = "CASC & TICC" if i == 0 else None
    plt.plot(latLong["Long"].values, latLong["Lat"].values, "*", markersize=4, color="red", label=label)

# s,e = 8176,8295
# latlong = dfLatLong[s:e+1]
# plt.plot(latLong["Long"].values, latLong["Lat"].values, "*", markersize=4, color="orange")

plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.legend()
plt.rcParams.update({'font.size': 14})
plt.savefig('carMap.eps', format='eps', bbox_inches='tight', dpi=1000)
plt.show()
```

```python
startB,endB = (27050, 27650)
# startA,endA = 5700, 5788
startA, endA = 8176,8295

def plotHeadingSpeed(s,e, plotFile):
    ll = (((dfLatLong.index > s)) & (dfLatLong.index <e))
    print(collapse(oldDf[ll]))
    fig, ax1 = plt.subplots()
    ts = df[ll]
    velVals = ts["Velocity"]
    rolling = velVals.rolling(window=10, min_periods=1, center=True).mean().values
    ax1.plot(ts.index/10.0, rolling, "b-")
    ax1.set_xlabel('Time (s)') 
    ax1.set_ylabel('Speed (km/h)', color='b')
    ax1.tick_params('y', colors='b')
    ax1.set_ylim(bottom=0, top=100)
    ax2 = ax1.twinx()
    ts = df[ll]
    headVals = ts["Heading"]
    rolling = headVals.rolling(window=10, min_periods=1, center=True).mean().values
    ax2.plot(ts.index/10.0,rolling, "r-")
    ax2.set_ylabel('Heading (degrees)', color='r')
    ax2.tick_params('y', colors='r')
    fig.tight_layout()
    plt.rcParams.update({'font.size': 14})
    plt.savefig(plotFile, format='eps', bbox_inches='tight', dpi=1000)
    plt.show()

plotHeadingSpeed(startA, endA, 'carHeadingSpeedA.eps')
plotHeadingSpeed(startB, endB, 'carHeadingSpeedB.eps')
```
